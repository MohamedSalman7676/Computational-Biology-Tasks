```python
import cobra
from cobra import Model, Reaction, Metabolite
#################################### Step_1: Add The Model ######################################
Model_1 = Model("First_Task")
#################################### Step_2: Add The Reactions ######################################
R1 = Reaction("r1")
R1.name = "r1"
R1.lower_bound = 0
R1.upper_bound = 1000

R2 = Reaction("r2")
R2.name = "r2"
R2.lower_bound = 0
R2.upper_bound = 1000

R3 = Reaction("r3")
R3.name = "r3"
R3.lower_bound = 0
R3.upper_bound = 1000

R4 = Reaction("r4")
R4.name = "r4"
R4.lower_bound = 1
R4.upper_bound = 1

R5 = Reaction("r5")
R5.name = "r5"
R5.lower_bound = .9
R5.upper_bound = .9

R6 = Reaction("r6")
R6.name = "r6"
R6.lower_bound = 0
R6.upper_bound = 1000
#################################### Step_3: Add The Metabolites ######################################
A = Metabolite("A", compartment="c")
B = Metabolite("B", compartment="c")
C = Metabolite("C", compartment="c")
D = Metabolite("D", compartment="c")
#################################### Step_4: Add Metabolites To Reactions ######################################
######################### Reaction_1: A ===> B ###########################
R1.add_metabolites({A: -1, B: 1})
######################### Reaction_2: B ===> C ###########################
R2.add_metabolites({B: -1, C: 1})
######################### Reaction_3:   ===> A ###########################
R3.add_metabolites({A: 1})
######################### Reaction_4: C ===>   ###########################
R4.add_metabolites({C: -1})
######################### Reaction_5: A ===> D ###########################
R5.add_metabolites({A: -1, D: 1})
######################### Reaction_6: D ===>   ###########################
R6.add_metabolites({D: -1})
#################################### Step_5: Reactions To The Model ######################################
Model_1.add_reactions([R1, R2, R3, R4, R5, R6])

```

```python
Model_1.objective = "r4"
```

```python
Model_1.summary()
```

```python
Model_1.optimize()
```
